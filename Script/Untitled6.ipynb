```python
import os
import json
import shutil

# ĐÃ SỬA LẠI ĐƯỜNG DẪN: Lùi sâu vào thêm 1 cấp thư mục
raw_img_dir = '/content/ICDAR2019/TrainImages/TrainImages'
raw_gt_dir = '/content/ICDAR2019/TrainGT/TrainGT'

# Thư mục đích của PaddleOCR
paddle_img_dir = '/content/PaddleOCR/train_data/images'
paddle_label_file = '/content/PaddleOCR/train_data/label.txt'

os.makedirs(paddle_img_dir, exist_ok=True)
count_japanese_images = 0

print("Bắt đầu quét 10.000 ảnh để lọc dữ liệu tiếng Nhật...")

with open(paddle_label_file, 'w', encoding='utf-8') as f_out:
    for gt_filename in os.listdir(raw_gt_dir):
        if not gt_filename.endswith('.txt'): continue

        gt_filepath = os.path.join(raw_gt_dir, gt_filename)

        # Khớp tên file: Đổi đuôi .txt thành .jpg
        img_filename = gt_filename.replace('.txt', '.jpg')

        if img_filename.startswith('gt_') and not os.path.exists(os.path.join(raw_img_dir, img_filename)):
            img_filename = img_filename.replace('gt_', '')

        img_filepath = os.path.join(raw_img_dir, img_filename)

        if not os.path.exists(img_filepath): continue

        labels = []
        has_japanese = False

        # Đọc từng dòng tọa độ trong file nhãn
        with open(gt_filepath, 'r', encoding='utf-8-sig') as f_in:
            lines = f_in.readlines()
            for line in lines:
                parts = line.strip().split(',')
                if len(parts) >= 10:
                    language = parts[8].strip()
                    text = ','.join(parts[9:]).strip()

                    # CHỈ LỌC TIẾNG NHẬT
                    if language.lower() in ['japanese', 'jpn']:
                        has_japanese = True
                        try:
                            pts = [
                                [int(parts[0]), int(parts[1])],
                                [int(parts[2]), int(parts[3])],
                                [int(parts[4]), int(parts[5])],
                                [int(parts[6]), int(parts[7])]
                            ]
                            labels.append({
                                "transcription": text,
                                "points": pts
                            })
                        except ValueError:
                            pass

        # Đưa ảnh có tiếng Nhật vào thư mục Train
        if has_japanese and len(labels) > 0:
            shutil.copy(img_filepath, os.path.join(paddle_img_dir, img_filename))
            label_str = json.dumps(labels, ensure_ascii=False)
            f_out.write(f"images/{img_filename}\t{label_str}\n")
            count_japanese_images += 1

print(f"✅ Xong! Đã gom thành công {count_japanese_images} bức ảnh tiếng Nhật đưa vào PaddleOCR.")
```
